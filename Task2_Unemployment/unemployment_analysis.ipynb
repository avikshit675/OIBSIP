# Unemployment Analysis in India

This project analyzes unemployment trends in India using data visualization techniques.

The dataset contains unemployment rates across different states and time periods.

Objective:
- Analyze unemployment trends in India
- Understand the impact of time and region on unemployment
- Visualize unemployment rates using graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("Unemployment in India.csv")
df2 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

df1.head()
```

```python
df1.info()
```

```python
df2.info()
```

```python
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

print(df1.columns)
print(df2.columns)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Region", y="Estimated Unemployment Rate (%)", data=df2)

plt.xticks(rotation=90)
plt.title("Unemployment Rate by Region")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(x="Date", y="Estimated Unemployment Rate (%)", data=df2)

plt.xticks(rotation=45)
plt.title("Unemployment Rate Over Time")
plt.show()
```

```python
pivot = df2.pivot_table(
    values="Estimated Unemployment Rate (%)",
    index="Region",
    columns="Frequency"
)

sns.heatmap(pivot, cmap="coolwarm", annot=True)

plt.title("Unemployment Heatmap")
plt.show()
```
